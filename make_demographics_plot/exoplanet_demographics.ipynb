```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('xtick', labelsize=45) 
matplotlib.rc('ytick', labelsize=45)

# use latex for font rendering
matplotlib.rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
matplotlib.rc('text', usetex=True)
```

```python
import pandas as pd

# Load the CSV file, skipping the first 35 rows and setting header as row 36
data = pd.read_csv('exoplanet_masses_periods.csv', skiprows=34, header=1)

# Now, df contains the data with column names from row 36 onwards
data
```

```python
# Create a boolean mask to identify rows where neither 'rv_flag' nor 'ast_flag' are equal to 1
mask = ((data['rv_flag'] == 1) | (data['ast_flag'] == 1))

# Use the mask to select the rows you want to keep
data_filtered = data[mask]

data_filtered
```

```python
print(data.columns)
```

```python
periods = data_filtered['pl_orbper']
masses = data_filtered['pl_bmasse']


```

```python
#from here: https://nssdc.gsfc.nasa.gov/planetary/factsheet/earthfact.html
m_earth = 1. # [m_earth]
p_earth = 365.256 # [days] sidereal
```

```python
#from here: https://nssdc.gsfc.nasa.gov/planetary/factsheet/jupiterfact.html
m_jup = 317.83 # [m_earth]
p_jup = 4332.589 # [days] sidereal
```

```python
#from here: https://nssdc.gsfc.nasa.gov/planetary/factsheet/saturnfact.html
m_saturn = 95.16 # [m_earth]
p_saturn = 10759.22 #[ days] sidereal
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

def getImage(path):
    return OffsetImage(plt.imread(path))
```

```python
fig, ax = plt.subplots(1, 1, figsize = [15,15])
ax.plot(periods, masses, 'o', color = "#7b3e65", alpha = 0.3, markersize = 13)


ax.text(.03, .97, str(len(periods)) + ' confirmed exoplanets around Sun-like stars',
       ha='left', va='top', transform=ax.transAxes, fontsize=27, fontweight='bold')

ax.set_yscale('log')
ax.set_xscale('log')




paths = [
    'earth.png',
    'jupiter.png',
    'saturn.png',
    ]

y = [m_earth,m_jup,m_saturn]
x = [p_earth,p_jup,p_saturn]

ax.scatter(x, y) 

for x0, y0, path in zip(x, y, paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), 
                        frameon=False)
    ax.add_artist(ab)
    

ax.set_xlim(.1, 500000)
ax.set_ylim(.1, 50000)

ax.set_ylabel(r'Mass [M$_{Earth}$]', fontsize = 36)
ax.set_xlabel('Orbital Period [days]', fontsize = 36)

fig.savefig('exoplanets_sun-like.pdf')

fig.tight_layout()
```

```python
fig, ax = plt.subplots(1, 1, figsize = [15,15])
ax.plot(periods, masses, 'o', color = "#7b3e65", alpha = 0.3, markersize = 13)


ax.text(.03, .97, r'\textbf{' + str(len(periods)) + ' exoplanets orbit Sun-like stars with}',
        ha='left', va='top', transform=ax.transAxes, fontsize=36, fontweight='bold')
ax.text(.03, .92, r'\textbf{dynamical masses from RVs or astrometry}',
        ha='left', va='top', transform=ax.transAxes, fontsize=36, fontweight='bold')


ax.set_yscale('log')
ax.set_xscale('log')


colors = ['#366537', 
          '#db372b', 
          '#b4943e']

planets = [
    'Earth',
    'Jupiter',
    'Saturn',
    ]

y = [m_earth,m_jup,m_saturn]
x = [p_earth,p_jup,p_saturn]

ax.scatter(x, y) 

for x0, y0, color, planet in zip(x, y, colors, planets):
    ax.plot(x0, y0, 'ks', color = color, markersize = 18, label=planet)
    

ax.set_xlim(.1, 500000)
ax.set_ylim(.1, 50000)

ax.set_ylabel(r'Mass [M$_{\oplus}$]', fontsize = 63)
ax.set_xlabel('Orbital Period [days]', fontsize = 63)

ax.legend(fontsize=36, loc=4)


fig.tight_layout()


fig.savefig('exoplanets_sun-like_paper.pdf')

```
